```python
from collections import defaultdict
```

```python
with open('pg16457.txt', 'r') as f:
    text = ' '.join(f.readlines())
```

```python
corpus = text.split(' ')
```

```python
len(corpus)
```

```python
def refactor(text):
    not_allowed = """\\-"_.•-—‘’',“”~@#[]%{}?;!^/&*()_=+\n"""
    return ''.join(str(c) for c in text if not not_allowed.__contains__(c))
```

```python
corpus = [refactor(text) for text in corpus]
```

```python
corpus = [text.lower() for text in corpus if len(text) > 0]
```

```python
corpus.sort()
```

```python
print(corpus)
```

```python
len(corpus)
```

```python
word_freq = defaultdict(int)
```

```python
for text in corpus:
    word_freq[text] += 1
```

```python
print(word_freq)
```

```python
alphabet = []
```

```python
for word in word_freq.keys():
    for letter in word:
        if letter not in alphabet:
            alphabet.append(letter)
alphabet.sort()
print(alphabet)
```

```python
splits = {word: [c for c in word] for word in word_freq.keys()}
```

```python
def compute_pair_freq(splits):
    pair_freq = defaultdict(int)
    for word, freq in word_freq.items():
        split = splits[word]
        if len(split) == 1:
            continue
        for i in range(len(split) - 1):
            pair = (split[i], split[i + 1])
            pair_freq[pair] += freq
    return pair_freq
```

```python
pair_freq = compute_pair_freq(splits)

for i, key in enumerate(pair_freq.keys()):
    print(f"{key}: {pair_freq[key]}")
    if i >= 500:
        break
```

```python
def get_best_pair(pair_freq) :
    best_pair = ""
    max_freq = None

    for pair, freq in pair_freq.items():
        if max_freq is None or max_freq < freq:
            best_pair = pair
            max_freq = freq
    return best_pair, max_freq
```

```python
def merge_pair(a, b, splits):
    for word in word_freq:
        split = splits[word]
        if len(split) == 1:
            continue

        i = 0
        while i < len(split) - 1:
            if split[i] == a and split[i + 1] == b:
                split = split[:i] + [a + b] + split[i + 2 :]
            else:
                i += 1
        splits[word] = split
    return splits
```
